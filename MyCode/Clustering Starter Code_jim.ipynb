# Part 3: Clustering using K-Means

Documentation on K Means clustering using sklearn in Python: https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

A discussion of 11 clustering algorithms with Python implementations is provided in: https://scikit-learn.org/stable/modules/clustering.html

```python
#K-means:
# 0. Assign random class to each point
#Repeat
# 1. Compute the mean of each class
# 2. Re-assign each point to a class using the closest mean
```

```python
print(imArr.shape)
k = 4

# Step 0:
nRows, nCols, nBands = imArr.shape
nPix = nRows*nCols
class_index = np.random.randint(low=0, high=k, size=nPix) # each pixel has a class
```

```python
# Step 1: Compute Class Means
class_means = np.zeros((k,nBands)) # a mean reflectance of each band of each class
for i in range(k):
    idx = np.where(class_index==i) # get the indexes of the pixels assigned to this class
    #spectra_for_this_class = np.squeeze(imList[idx,:])
    spectra_for_this_class = imList[idx] # get the pixels (and spectra) that correspond to these indexes (and this class)
    #class_means[i,:] = np.mean(spectra_for_this_class, axis=0)
    class_means[i] = np.mean(spectra_for_this_class, axis=0) # go through the rows/pixels/spectra (collapse the rows) and get a mean for each band; each class has a mean reflectance for each band

print(imList.shape)
print(spectra_for_this_class.shape)
print(class_means.shape)
```

```python
# Step 2: Re-assign Class Labels
distance_to_mean = np.zeros((nPix,k)) # each pixel's distance (as a point, considering all bands) to the mean of each class (as a point, considering all bands)
for i_cls in range(k):
    imListTemp = copy.copy(imList)
    for j in range(nBands):
        imListTemp[:,j] = imListTemp[:,j] - class_means[i_cls,j] # get the difference between each pixel's jth band's reflectance and the mean reflectance of that band of this class
    # imListTemp now has has set of difference values for each pixel (with respect to the present class mean)
    
    # note that we have k classes: k means (points), one for each class
    distance_to_mean[:,i_cls] = np.sum(imListTemp**2, axis=1) # sum up the squares of all of the difference values for each pixel accross all bands, thus collapsing the cols; you could take the square root of the sum for the actual distance
        # this gives you each pixel's distance (as a point, considering all bands) to the mean of this (the present) class (as a point, considering all bands)
#now distance_to_mean has k distance values for each pixel
class_index = np.argmin(distance_to_mean, axis=1)
    # The numpy.argmin() method returns indices of the min element of the array in a particular axis.
    # Get the index (class) of the smallest distance across the k classes for each pixel, collapsing the columns (classes)
    # recall that each pixel/spectrum has only one class
```

```python
plt.figure(figsize=(15,5))
colors = ['r','g','b','y']
for i in range(k): 
    idx = np.where(class_index==i)
    #S = np.squeeze(imList[idx,:])
    S = imList[idx]
    plt.scatter(S[:,i1], S[:,i2], c=colors[i], s=5, alpha=0.5, label=f'Class {i}');
plt.xlim(0,8000)
plt.grid(True)
plt.xlabel(f'Reflectance at {wl[i1]:.1f} nm')
plt.ylabel(f'Reflectance at {wl[i2]:.1f} nm')
plt.title('Scatterplot of the Washington DC Image');
plt.legend()
```
